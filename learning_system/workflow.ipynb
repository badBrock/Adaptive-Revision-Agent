```python
# Complete Workflow Visualization Script
# Copy and paste this entire code block into one Jupyter cell

from IPython.display import Image, display
import sys
import os

# Add current directory to path (adjust if needed)
sys.path.append('.')
sys.path.append('./tools')

def visualize_all_workflows():
    """Complete workflow visualizer that handles all cases"""
    
    print("üéì ADAPTIVE TUTORING SYSTEM - WORKFLOW VISUALIZATION")
    print("=" * 80)
    print("Generated on: Wednesday, September 10, 2025, 4:36 AM IST")
    print("=" * 80)
    
    # Workflow configurations
    workflows = [
        {
            "module": "boss_agent",
            "function": "create_boss_agent_workflow", 
            "name": "Boss Agent",
            "description": "Orchestrates learning sessions and routes between agents",
            "emoji": "ü§ñ"
        },
        {
            "module": "quiz_agent",
            "function": "create_quiz_agent_workflow",
            "name": "Quiz Agent", 
            "description": "Conducts Q&A sessions and provides assessment",
            "emoji": "üéØ"
        },
        {
            "module": "teaching_agent",
            "function": "create_teaching_agent_workflow",
            "name": "Teaching Agent",
            "description": "Implements Feynman Technique for deep learning", 
            "emoji": "üéì"
        }
    ]
    
    successful = 0
    failed = []
    
    for config in workflows:
        success = visualize_single_workflow(config)
        if success:
            successful += 1
        else:
            failed.append(config["name"])
    
    # Final summary
    print(f"\n{'='*80}")
    print(f"üìä VISUALIZATION SUMMARY")
    print(f"{'='*80}")
    print(f"‚úÖ Successfully visualized: {successful}/{len(workflows)} workflows")
    
    if failed:
        print(f"‚ùå Failed workflows: {', '.join(failed)}")
        print(f"üí° Tip: Make sure all .py files are in the current directory")
    
    if successful > 0:
        print(f"\nüéâ Visualization complete! Check the graphs above.")
    
    return successful, failed

def visualize_single_workflow(config):
    """Visualize a single workflow - handles all edge cases"""
    try:
        print(f"\n{config['emoji']} {config['name'].upper()} WORKFLOW")
        print("=" * 60)
        print(f"üìã Purpose: {config['description']}")
        
        # Step 1: Import module
        try:
            module = __import__(config["module"])
            print(f"‚úÖ Module '{config['module']}' imported successfully")
        except ImportError as e:
            print(f"‚ùå Cannot import module '{config['module']}': {e}")
            return False
        
        # Step 2: Get workflow creation function/object
        try:
            workflow_creator = getattr(module, config["function"])
            print(f"‚úÖ Found function '{config['function']}'")
        except AttributeError as e:
            print(f"‚ùå Function '{config['function']}' not found in module: {e}")
            return False
        
        # Step 3: Handle different object types
        workflow_obj = None
        
        # Case 1: It's a callable function
        if callable(workflow_creator):
            try:
                print("üì¶ Calling workflow creation function...")
                workflow_obj = workflow_creator()
                print("‚úÖ Workflow created from function")
            except Exception as e:
                print(f"‚ùå Failed to call workflow function: {e}")
                return False
        
        # Case 2: It's already a compiled workflow object
        else:
            print("üì¶ Using pre-compiled workflow object...")
            workflow_obj = workflow_creator
        
        # Step 4: Verify workflow object has required methods
        if not hasattr(workflow_obj, 'get_graph'):
            print(f"‚ùå Workflow object missing 'get_graph' method")
            print(f"   Object type: {type(workflow_obj)}")
            available_methods = [m for m in dir(workflow_obj) if not m.startswith('_')]
            print(f"   Available methods: {available_methods[:10]}...")  # Show first 10
            return False
        
        # Step 5: Generate graph
        try:
            print("üé® Generating workflow graph...")
            graph = workflow_obj.get_graph(xray=True)
            print("‚úÖ Graph generated successfully")
        except Exception as e:
            print(f"‚ùå Failed to generate graph: {e}")
            return False
        
        # Step 6: Render as PNG
        try:
            print("üñºÔ∏è Rendering Mermaid PNG...")
            png_data = graph.draw_mermaid_png()
            print("‚úÖ PNG rendered successfully")
        except Exception as e:
            print(f"‚ùå Failed to render PNG: {e}")
            print("üí° Note: This might be due to missing Mermaid dependencies")
            return False
        
        # Step 7: Display the visualization
        try:
            display(Image(png_data))
            print("‚úÖ Workflow visualization displayed!")
        except Exception as e:
            print(f"‚ùå Failed to display image: {e}")
            return False
        
        # Step 8: Show workflow statistics
        try:
            nodes = list(graph.nodes())
            edges = list(graph.edges())
            print(f"\nüìä Workflow Statistics:")
            print(f"   ‚Ä¢ Total Nodes: {len(nodes)}")
            print(f"   ‚Ä¢ Total Edges: {len(edges)}")
            print(f"   ‚Ä¢ Node Names: {nodes}")
            print(f"   ‚Ä¢ Workflow Type: {type(workflow_obj).__name__}")
        except Exception as e:
            print(f"‚ö†Ô∏è Could not get workflow stats: {e}")
        
        return True
        
    except Exception as e:
        print(f"‚ùå Unexpected error visualizing {config['name']}: {e}")
        import traceback
        print("üîç Full error traceback:")
        traceback.print_exc()
        return False

def show_integration_overview():
    """Display system integration overview"""
    print(f"\n{'='*80}")
    print(f"üîÑ SYSTEM INTEGRATION OVERVIEW")
    print(f"{'='*80}")
    
    integration_flow = """
    ‚îå‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îê
    ‚îÇ   Boss Agent    ‚îÇ ‚Üê‚îÄ‚îÄ User Request + Content Path
    ‚îÇ  (Orchestrator) ‚îÇ
    ‚îî‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚î¨‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îò
              ‚îÇ (Routes based on user performance)
              ‚ñº
    ‚îå‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îê       ‚îå‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îê
    ‚îÇ   Quiz Agent    ‚îÇ       ‚îÇ Teaching Agent  ‚îÇ
    ‚îÇ  (Assessment)   ‚îÇ ‚Üê‚îÄ‚îÄ‚îÄ‚Üí ‚îÇ (Explanation)   ‚îÇ
    ‚îÇ                 ‚îÇ       ‚îÇ                 ‚îÇ
    ‚îî‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚î¨‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îò       ‚îî‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚î¨‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îò
              ‚îÇ                         ‚îÇ
              ‚ñº                         ‚ñº
    ‚îå‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îê
    ‚îÇ     Meta-Learning & Progress Tracking           ‚îÇ
    ‚îÇ   (Adaptive recommendations & state updates)    ‚îÇ
    ‚îî‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚î¨‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îò
                      ‚îÇ
                      ‚ñº
    ‚îå‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îê
    ‚îÇ   Boss Agent    ‚îÇ ‚Üê‚îÄ‚îÄ Next session routing
    ‚îÇ  (Re-routing)   ‚îÇ
    ‚îî‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îò
    """
    
    print(integration_flow)
    
    print(f"\nüìà KEY WORKFLOW PATTERNS:")
    print(f"   1. Boss Agent: Initialize ‚Üí Discover Content ‚Üí Reasoning ‚Üí Execute ‚Üí Update")
    print(f"   2. Quiz Agent: Load Content ‚Üí Plan Questions ‚Üí Q&A Loop ‚Üí Score ‚Üí Finalize")
    print(f"   3. Teaching Agent: Load Content ‚Üí Generate Prompts ‚Üí Teaching Loop ‚Üí Evaluate")
    print(f"   4. Shared Components: ContentCache, StateManager, MetaLearning")

# Main execution
if __name__ == "__main__":
    # Run complete visualization
    successful_count, failed_workflows = visualize_all_workflows()
    
    # Show integration overview
    show_integration_overview()
    
    # Final status message
    if successful_count > 0:
        print(f"\nüéâ SUCCESS! {successful_count} workflow(s) visualized.")
        print(f"üìä Check the Mermaid diagrams above to see your agent flows!")
    else:
        print(f"\n‚ö†Ô∏è No workflows could be visualized.")
        print(f"üîç Make sure your .py files are in the current directory.")
        print(f"üí° Try running: import boss_agent, quiz_agent, teaching_agent")

# Execute the visualization
visualize_all_workflows()
show_integration_overview()
```
