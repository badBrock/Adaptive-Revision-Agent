```python
# Complete Workflow Visualization Script
# Copy and paste this entire code block into one Jupyter cell

from IPython.display import Image, display
import sys
import os

# Add current directory to path (adjust if needed)
sys.path.append('.')
sys.path.append('./tools')

def visualize_all_workflows():
    """Complete workflow visualizer that handles all cases"""
    
    print("🎓 ADAPTIVE TUTORING SYSTEM - WORKFLOW VISUALIZATION")
    print("=" * 80)
    print("Generated on: Wednesday, September 10, 2025, 4:36 AM IST")
    print("=" * 80)
    
    # Workflow configurations
    workflows = [
        {
            "module": "boss_agent",
            "function": "create_boss_agent_workflow", 
            "name": "Boss Agent",
            "description": "Orchestrates learning sessions and routes between agents",
            "emoji": "🤖"
        },
        {
            "module": "quiz_agent",
            "function": "create_quiz_agent_workflow",
            "name": "Quiz Agent", 
            "description": "Conducts Q&A sessions and provides assessment",
            "emoji": "🎯"
        },
        {
            "module": "teaching_agent",
            "function": "create_teaching_agent_workflow",
            "name": "Teaching Agent",
            "description": "Implements Feynman Technique for deep learning", 
            "emoji": "🎓"
        }
    ]
    
    successful = 0
    failed = []
    
    for config in workflows:
        success = visualize_single_workflow(config)
        if success:
            successful += 1
        else:
            failed.append(config["name"])
    
    # Final summary
    print(f"\n{'='*80}")
    print(f"📊 VISUALIZATION SUMMARY")
    print(f"{'='*80}")
    print(f"✅ Successfully visualized: {successful}/{len(workflows)} workflows")
    
    if failed:
        print(f"❌ Failed workflows: {', '.join(failed)}")
        print(f"💡 Tip: Make sure all .py files are in the current directory")
    
    if successful > 0:
        print(f"\n🎉 Visualization complete! Check the graphs above.")
    
    return successful, failed

def visualize_single_workflow(config):
    """Visualize a single workflow - handles all edge cases"""
    try:
        print(f"\n{config['emoji']} {config['name'].upper()} WORKFLOW")
        print("=" * 60)
        print(f"📋 Purpose: {config['description']}")
        
        # Step 1: Import module
        try:
            module = __import__(config["module"])
            print(f"✅ Module '{config['module']}' imported successfully")
        except ImportError as e:
            print(f"❌ Cannot import module '{config['module']}': {e}")
            return False
        
        # Step 2: Get workflow creation function/object
        try:
            workflow_creator = getattr(module, config["function"])
            print(f"✅ Found function '{config['function']}'")
        except AttributeError as e:
            print(f"❌ Function '{config['function']}' not found in module: {e}")
            return False
        
        # Step 3: Handle different object types
        workflow_obj = None
        
        # Case 1: It's a callable function
        if callable(workflow_creator):
            try:
                print("📦 Calling workflow creation function...")
                workflow_obj = workflow_creator()
                print("✅ Workflow created from function")
            except Exception as e:
                print(f"❌ Failed to call workflow function: {e}")
                return False
        
        # Case 2: It's already a compiled workflow object
        else:
            print("📦 Using pre-compiled workflow object...")
            workflow_obj = workflow_creator
        
        # Step 4: Verify workflow object has required methods
        if not hasattr(workflow_obj, 'get_graph'):
            print(f"❌ Workflow object missing 'get_graph' method")
            print(f"   Object type: {type(workflow_obj)}")
            available_methods = [m for m in dir(workflow_obj) if not m.startswith('_')]
            print(f"   Available methods: {available_methods[:10]}...")  # Show first 10
            return False
        
        # Step 5: Generate graph
        try:
            print("🎨 Generating workflow graph...")
            graph = workflow_obj.get_graph(xray=True)
            print("✅ Graph generated successfully")
        except Exception as e:
            print(f"❌ Failed to generate graph: {e}")
            return False
        
        # Step 6: Render as PNG
        try:
            print("🖼️ Rendering Mermaid PNG...")
            png_data = graph.draw_mermaid_png()
            print("✅ PNG rendered successfully")
        except Exception as e:
            print(f"❌ Failed to render PNG: {e}")
            print("💡 Note: This might be due to missing Mermaid dependencies")
            return False
        
        # Step 7: Display the visualization
        try:
            display(Image(png_data))
            print("✅ Workflow visualization displayed!")
        except Exception as e:
            print(f"❌ Failed to display image: {e}")
            return False
        
        # Step 8: Show workflow statistics
        try:
            nodes = list(graph.nodes())
            edges = list(graph.edges())
            print(f"\n📊 Workflow Statistics:")
            print(f"   • Total Nodes: {len(nodes)}")
            print(f"   • Total Edges: {len(edges)}")
            print(f"   • Node Names: {nodes}")
            print(f"   • Workflow Type: {type(workflow_obj).__name__}")
        except Exception as e:
            print(f"⚠️ Could not get workflow stats: {e}")
        
        return True
        
    except Exception as e:
        print(f"❌ Unexpected error visualizing {config['name']}: {e}")
        import traceback
        print("🔍 Full error traceback:")
        traceback.print_exc()
        return False

def show_integration_overview():
    """Display system integration overview"""
    print(f"\n{'='*80}")
    print(f"🔄 SYSTEM INTEGRATION OVERVIEW")
    print(f"{'='*80}")
    
    integration_flow = """
    ┌─────────────────┐
    │   Boss Agent    │ ←── User Request + Content Path
    │  (Orchestrator) │
    └─────────┬───────┘
              │ (Routes based on user performance)
              ▼
    ┌─────────────────┐       ┌─────────────────┐
    │   Quiz Agent    │       │ Teaching Agent  │
    │  (Assessment)   │ ←───→ │ (Explanation)   │
    │                 │       │                 │
    └─────────┬───────┘       └─────────┬───────┘
              │                         │
              ▼                         ▼
    ┌─────────────────────────────────────────────────┐
    │     Meta-Learning & Progress Tracking           │
    │   (Adaptive recommendations & state updates)    │
    └─────────────────┬───────────────────────────────┘
                      │
                      ▼
    ┌─────────────────┐
    │   Boss Agent    │ ←── Next session routing
    │  (Re-routing)   │
    └─────────────────┘
    """
    
    print(integration_flow)
    
    print(f"\n📈 KEY WORKFLOW PATTERNS:")
    print(f"   1. Boss Agent: Initialize → Discover Content → Reasoning → Execute → Update")
    print(f"   2. Quiz Agent: Load Content → Plan Questions → Q&A Loop → Score → Finalize")
    print(f"   3. Teaching Agent: Load Content → Generate Prompts → Teaching Loop → Evaluate")
    print(f"   4. Shared Components: ContentCache, StateManager, MetaLearning")

# Main execution
if __name__ == "__main__":
    # Run complete visualization
    successful_count, failed_workflows = visualize_all_workflows()
    
    # Show integration overview
    show_integration_overview()
    
    # Final status message
    if successful_count > 0:
        print(f"\n🎉 SUCCESS! {successful_count} workflow(s) visualized.")
        print(f"📊 Check the Mermaid diagrams above to see your agent flows!")
    else:
        print(f"\n⚠️ No workflows could be visualized.")
        print(f"🔍 Make sure your .py files are in the current directory.")
        print(f"💡 Try running: import boss_agent, quiz_agent, teaching_agent")

# Execute the visualization
visualize_all_workflows()
show_integration_overview()
```
